## Algo or Example

Zooming can be viewed as oversampling and shrinking can be viewed as undersampling. Both the operations are applied on image processing.
Zooming requires two processes:-
Firstly, the creation of new pixels.
Then fill or assignment of gray levels to these new pixels.
The spacing in the grid is less than 1 pixel because we are fitting it over a smaller image.

Algorithm for **zooming** is as follows:-
1. We will take an image of m*n.
2. Then we zoom the image by km*kn, here k is the zooming factor.
3. Then  replicate the values of each pixels by k times. 

In zooming as we have zooming factor like 2, then it will replicate each pixel value by 2. 


Algorithm for **shrinking** of image is as follows:-
1. We will take an image of m*n.
2. Then we will shrink the image by (m/k)*(n/k).
3. Then fill the values of each pixel.

In shrinking , suppose we have factor as 2 then it will skip the alternate rows and columns for decreasing the size of image.

Advantages of this technique is that it is very simple . We just have to copy pixels.
Disadvantage is that :-
This technique gives blur image as an output. As we increase the zooming factor value, the image gets bigger and bigger and become more blur.

```python
# We have image like
print('We have gray scale image and it is like : \n\n')

emp_image = np.random.randint(1,255,size = (6,6))

print('Image original shape is  : {}'.format(emp_image.shape))
print('\nImage is :\n\n {} '.format(emp_image))

# Shrinking and Zooming factor is '2'
factor = 2
Height, Width = emp_image.shape

# Shrink image 

Shrink_exp = np.ones((int(emp_image.shape[0]/factor), int(emp_image.shape[1]/factor)), int)

print('\nImage Dimensions After Shrink : ',Shrink_exp.shape)
for height in range(int(Height/factor)):
    for width in range(int(Width/factor)):
        emp_ploc = np.floor(height*factor)
        emp_qloc = np.floor(width*factor)
        Shrink_exp[height , width] = emp_image[int(emp_ploc), int(emp_qloc)]

print('\n', Shrink_exp)

# Zoomed image

exp_img = np.ones((emp_image.shape[0]*factor, emp_image.shape[1]*factor), int)

print('\nImage Dimensions After Zooming :  ', exp_img.shape)

pixelScaleW = Width/ exp_img.shape[1]
pixelScaleH = Height/ exp_img.shape[0]

for height in range(int(Height*factor) ):
    for width in range(int(Width*factor) ):
        exp_img[width , height] = emp_image[int(height * pixelScaleH), int(width*pixelScaleW)]

print('\n', exp_img)

```

### Optimize way of image shrinking

```python
## optimize way of image shrinking using numpy 
# Factor is '2'
Factor = 2

Shrink_img = emp_image[::factor, ::factor]
print('Oringial Image :\n\n', emp_image )
print('\nOriginal Image shape : ', emp_image.shape)

print('\nShring Image :\n\n', Shrink_img)
print('\nShrink Image shape : ', Shrink_img.shape)
```

## Part- A 

# Shrinking Image

```python
# Taking Input from user for how much the image gets shrink 
# Input is always integer and lies between (1- 10)
Factor = int(input('Enter the Shrinking factor : '))
```

```python
# Original Image shape
Image = cv2.imread('assets/img.jpg', 0)
print('Original Image shape : {}\n'.format(Image.shape))
plt.imshow(Image, cmap='gray')
plt.show()

# After Shrinking shape is
print('After Shrinking, Image shape is: {} \n'.format((int(Image.shape[0]/Factor), int(Image.shape[1]/Factor))))
```

```python
# First find Height and width of original image. 
Height, Width = Image.shape

# Create New Empty image with shrink factor

# New height 
new_height = Image.shape[0]/Factor
 
# New width
new_width = Image.shape[1]/Factor

# After finding new height and width change in integer because height and width are always in integer
new_height = int(new_height)
new_width = int(new_width)                

# Using numpy funtion 'np.ones()'
Shrink_img = np.ones((new_height, new_width))

print('\n New Image shape : ',Shrink_img.shape)
```

```python
#Shrinking Algo
for height in range(new_height):
    for width in range(new_width):
        Hpixel = np.floor(height*Factor)
        Wpixel = np.floor(width*Factor)
        Shrink_img[height , width] = Image[int(Hpixel), int(Wpixel)]

Img_path = 'Output/shrin_imgs.jpg'
cv2.imwrite(Img_path, Shrink_img)

# Result_img = cv2.imread(shrin_imgs)
plt.imshow(Shrink_img, cmap='gray')
plt.show()
```

# Zooming Image

```python
# Taking Input from user for how much he zoom the image 
# Input is always integer and lies between (1- 10)
Factor = int(input('Enter the zooming factor : '))
```

```python
# Original Image shape
Image = cv2.imread('assets/img.jpg', 0)
print('Original Image shape : {}\n'.format(Image.shape))
plt.imshow(Image, cmap='gray')
plt.show()

# After Zooming Image shape is:
print('After Zoom Image shape : {} \n'.format((int(Image.shape[0]*Factor), int(Image.shape[1]*Factor))))
```

```python
# First find original image Height and width 
Height, Width = Image.shape

# Create New Empty image with Zooming factor

# New height 
new_height = Image.shape[0]*Factor
 
# New width
new_width = Image.shape[1]*Factor

# After find new height and width change in integer because height and width are always integer
new_height = int(new_height)
new_width = int(new_width)                

# Using numpy funtion 'np.ones()'
Zoom_img = np.ones((new_height, new_width))

print('\n New Image shape : ',Zoom_img.shape)
```

```python
# Zooming the image
Wpixel = Width/ Zoom_img.shape[1]
Hpixel = Height/ Zoom_img.shape[0]
for height in range(int(Height*Factor)):
    for width in range(int(Width*Factor)):
        Zoom_img[height , width] = Image[int(height * Hpixel), int(width*Wpixel)]
# Save image on given path
Img_path = 'Output/Zoom_img.jpg'
cv2.imwrite(Img_path, Zoom_img)

# Display Image
plt.imshow(Zoom_img, cmap='gray')
plt.show()
```
